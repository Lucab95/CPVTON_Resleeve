```python
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
initial_path=os.getcwd()
os.chdir(Path(initial_path).parent) # change current dir to the parent directory
from src.visualization.visualize import read_image,show_raw_images, show_images

txt_path = Path('data/wild_pairs.txt')
raw_folder = Path('data/raw')
```

```python
images = []
for line in txt_path.open():
    line=line.strip()
    img_path = os.path.join(raw_folder,"image",line.split(' ')[0])
    cloth_path= os.path.join(raw_folder,"cloth",line.split(' ')[1])
    images.append((img_path,cloth_path))
```

```python
n_visualizations = 5
```

```python
titles=[]
shows = []
module = 'PIL'
size = (400,400)
for i in range(len(images)):
    img_path,cloth_path = images[i]
    titles.append(Path(img_path).parts[-1].replace('.jpg',''))
    shows.append((read_image(img_path, module, size=size)))
    titles.append(Path(cloth_path).parts[-1].replace('.jpg',''))
    shows.append((read_image(cloth_path, module, size=size)))
```
